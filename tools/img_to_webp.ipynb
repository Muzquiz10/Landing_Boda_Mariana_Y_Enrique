# Transformar sólo una imagen a webp

```python
from PIL import Image
import os

def convertir_a_webp(input_path, calidad=80):
    """
    Convierte una imagen PNG/JPG a formato WebP.
    """
    nombre, _ = os.path.splitext(input_path)
    output_path = f"{nombre}.webp"

    with Image.open(input_path) as img:
        img.save(output_path, "WEBP", quality=calidad)

    return output_path


# Ejemplo de uso
ruta = "../assets/IMG/portada.jpg"  
resultado = convertir_a_webp(ruta)
print("Imagen convertida:", resultado)
```

# Transformar a Webp de manera masiva

```python
from PIL import Image
import os

def convertir_carpeta_a_webp(carpeta, calidad=80):
    """
    Convierte todas las imágenes PNG/JPG/JPEG de una carpeta a WebP.
    """
    extensiones = (".png", ".jpg", ".jpeg")

    for archivo in os.listdir(carpeta):
        if archivo.lower().endswith(extensiones):
            ruta = os.path.join(carpeta, archivo)
            nombre, _ = os.path.splitext(ruta)
            salida = f"{nombre}.webp"

            with Image.open(ruta) as img:
                img.save(salida, "WEBP", quality=calidad)

            print(f"Convertido: {archivo} → {os.path.basename(salida)}")


# Ejemplo de uso
convertir_carpeta_a_webp("../assets/IMG/")
```
