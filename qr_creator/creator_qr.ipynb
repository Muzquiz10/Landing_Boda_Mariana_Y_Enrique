### Creador QR

```python
# Carpeta donde guardar los QR
output_dir = "qrs"
os.makedirs(output_dir, exist_ok=True)

# Cargar logo
logo = Image.open("logo_qr.png").convert("RGBA")

for _, row in df.iterrows():
    enlace = str(row["Enlace landing"]).strip()
    nombre = str(row["Nombre"]).strip()

    if enlace == "" or enlace.lower() == "nan":
        continue

    nombre_archivo = nombre.replace(" ", "_").replace("/", "_")

    # --- Crear QR ---
    qr = qrcode.QRCode(
        version=None,
        error_correction=ERROR_CORRECT_H,
        box_size=10,
        border=4,
    )

    qr.add_data(enlace)
    qr.make(fit=True)

    qr_img = qr.make_image(fill_color="#ae8d76", back_color="white").convert("RGBA")

    qr_width, qr_height = qr_img.size

    # --- Tamaño del logo (20–22% recomendado) ---
    logo_size = int(qr_width * 0.22)
    logo_resized = logo.resize((logo_size, logo_size), Image.LANCZOS)

    # --- Coordenadas centradas ---
    x = (qr_width - logo_size) // 2
    y = (qr_height - logo_size) // 2

    # --- 1️⃣ Crear hueco blanco en el centro ---
    draw = ImageDraw.Draw(qr_img)
    padding = int(logo_size * 0.15)  # margen alrededor del logo

    draw.ellipse(
        [
            x - padding,
            y - padding,
            x + logo_size + padding,
            y + logo_size + padding
        ],
        fill="white"
    )

    # --- 2️⃣ Pegar logo ---
    qr_img.paste(logo_resized, (x, y), logo_resized)

    # --- Guardar ---
    qr_img.save(f"{output_dir}/{nombre_archivo}.png")


## para mejorar el qr: https://chatgpt.com/c/693f2d8a-7088-8331-8350-cb3850df2227
```
